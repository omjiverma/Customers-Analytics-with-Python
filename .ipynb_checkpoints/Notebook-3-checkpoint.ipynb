# Purchase Predictive Analysis

## Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="whitegrid", rc=custom_params,palette='Set1')
from sklearn.linear_model import LogisticRegression 
import warnings
warnings.filterwarnings('ignore')
```

## Load Data

```python
df_purchase = pd.read_csv('purchase_segmented.csv',index_col=0)
```

```python
df_purchase.head()
```

## Data Preparation

```python
df_purchase_predictors = df_purchase.copy()
```

```python
segment_dummies = pd.get_dummies(df_purchase_predictors['Segment'], prefix = 'Segment', prefix_sep = '_')
df_purchase_predictors = pd.concat([df_purchase_predictors, segment_dummies], axis = 1)
```

## Purchase Probability Model

```python
y = df_purchase_predictors['Incidence']
```

```python
X = df_purchase_predictors[['Price_1','Price_2','Price_3','Price_4','Price_5']].mean(axis=1)
X = pd.DataFrame(X,columns=['Price_mean'])
```

```python
lr_purchase = LogisticRegression(solver = 'sag')
lr_purchase.fit(X, y)
```

```python
beta = lr_purchase.coef_[0,0]
```

## Price Elasticity of Purchase Probability

```python
# Generating price sequence from 0.5 to 3.0
price_range = np.arange(0.5, 3.0, 0.01)
df_price_range = pd.DataFrame(price_range,columns=['Price_range'])
```

```python
# Purchase Porbability prediction
purchase_proba = lr_purchase.predict_proba(price_range.reshape(-1,1))[:,1]
```

```python
#price elasticity
pe = beta * (df_price_range['Price_range'] * (1- purchase_proba))
df_pe = df_price_range
df_pe['Mean_PE'] = pe
```

```python
sns.lineplot(data=df_pe,x='Price_range',y='Mean_PE')
plt.title('Price Elasiticity of Purchase Probability',fontsize=16)
plt.ylabel('Elasticity');
```

* More Customers prefers to buy at lower price

## Purchase Probability by Segment

### Segment 1

```python
segment_1 = df_purchase_predictors[df_purchase_predictors['Segment'] == 1]
```

```python
y1 = segment_1['Incidence']
X1 = segment_1[['Price_1','Price_2','Price_3','Price_4','Price_5']].mean(axis=1)
X1 = pd.DataFrame(X1,columns=['Price_mean'])
```

```python
lr_segment_1 = LogisticRegression(solver = 'sag')
lr_segment_1.fit(X1, y1)
```

```python
beta1 =lr_segment_1.coef_[0,0]
```

```python
purchase_segment1_proba = lr_segment_1.predict_proba(price_range.reshape(-1,1))[:,1]
pe_1 = beta1 * (price_range * (1- purchase_segment1_proba))
df_pe_1 = pd.DataFrame(price_range,columns=['Price_range'])
df_pe_1['Mean_PE'] = pe_1
```

```python
df_segment_pe = df_pe_1
df_segment_pe = df_segment_pe.set_index('Price_range')
df_segment_pe.rename(columns={'Mean_PE':'Segment_1'},inplace=True)
```

```python
df_segment_pe.plot(figsize=(10,6))
plt.title('Price Elasticity of Purchase Probabilities');
```
